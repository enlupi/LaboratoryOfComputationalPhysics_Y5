```python

def PrimeNumber(N):
    mask = np.array([True for i in range(N+1)])
    n  = 2
    while n < N+1:
        i = 2
        while i*n < N+1:
            mask[i*n] = False
            i += 1
        n +=1
    a = np.linspace(0, N, N+1)
    return a[mask]

def Sieve(N):
    mask = np.array([True for i in range(N+1)])
    n = 2
    while n*n < N+1:
        if mask[n] == True:
            for i in range(n*n, N+1, n):
                mask[i] = False
        n += 1
    a = np.linspace(0, N, N+1)
    return a[mask]

print(PrimeNumber(30))
print(Sieve(30))
```

```python
from scipy import linalg as la
```

```python
N = 1000
X = np.zeros((3, N))
X[0] = npr.normal(loc=0, scale=1, size=N)
X[1] = npr.normal(loc=0, scale=3, size=N)+X[0]
X[2] = 2*X[0]+X[1]
```

```python
cov = np.cov(X)
evl, evt = la.eig(cov)
perm = np.flip(np.argsort(evl))
evl=np.real_if_close(evl[perm])
evt=evt[:,perm]
print(f"Eigenvalues:{evl}")
print(f"Eigenvectors:\n{evt}")
```

```python
U, spectrum, Vt = la.svd(X)
print(f"Eigenvalues for SVD:{spectrum**2/(N-1)}")
print(f"Eigenvectors for SVD:\n{U}")

print("Are all eigenvalues the same?",  np.allclose(evl, spectrum**2/(N-1), rtol=1e-3))
print("Are all eigenvectors the same?", np.allclose(U, evt, rtol=1e-3))
```

```python
Tr = cov.trace()
var = evl/Tr
for l, v in zip(evl, var):
    print(f"To eigenvalue {l} corresponds {v*100}% variability")

Xr = np.dot(evt.T, X)
X_reduced = Xr[0:1,]

print(f"X_reduced contains {var[0] + var[1]} variability")
```

```python
# For each camera we randomly select a pair (theta, phi)
theta = np.random.random(n)*np.pi
phi = np.random.random(n)*2*np.pi

z=np.array([0,0,1])
v=np.array([np.sin(theta)*np.cos(phi), np.sin(theta)*np.sin(phi), np.cos(theta)])
ax=np.cross(z, v.T)

print(ax)
```

```python
from numpy import cross, eye, dot
from scipy.linalg import expm, norm

def RotM(axis, theta):
    return expm(cross(eye(3), axis/norm(axis)*theta))

for i in range(0, n):
    M = RotM(ax[i], -theta[i])
    for j in range(N):
        data[3+i*3:6+i*3, j]=np.dot(M, data[0:3, j])
```

```python
l, U = la.eig(np.cov(data))
perm = np.flip(np.argsort(l))
l=np.real_if_close(l[perm])
U=U[:,perm]
Tr = np.sum(l)
var = l/Tr
for l, v in zip(l, var):
    print(f"To eigenvalue {l} corresponds {v*100}% variability")

print(f"X_reduced contains {var[0] + var[1]} variability")
```

```python
hl = 3.052*60 #sec
N = 1000
Ttot = round(5*hl)

def prob(t, T=hl):
    return 1 - 2**(-t/T)

NTl = np.zeros(Ttot)
NPb = np.zeros(Ttot)

NTl[0] = N

p = npr.uniform(0, 1, N)
for t in range(1, Ttot):
    NPb[t] = np.sum(p < prob(t))
    NTl[t] = NTl[0] - NPb[t] 
    
time = np.arange(0, Ttot, 1)
fig = plt.figure(figsize=(8,5))
plt.plot(time, NTl, label="N Tl")
plt.plot(time, NPb, label="N Pb")
plt.axvline(hl, ls="--", label="Half life")
plt.axhline(N/2, ls="--", color="green", label="Half initial N")
plt.legend(loc="best")
plt.show()
```

```python
tau = hl/np.log(2)
def inv_exp(z, l=1/tau):
    return -np.log(1-z)/l

z = npr.uniform(0, 1, N)
t = sorted(inv_exp(z))
NTl = np.array([N-i for i in range(N)])

plt.plot(t, NTl, label="N Tl")
plt.axvline(hl, ls="--", label="half life")
plt.legend()
```
